```python
import pandas as pd

# Read CSV with specified encoding
df_raw = pd.read_csv('raw_data.csv', encoding='utf-8')

# Combine filtering conditions
df = df_raw[df_raw['pricing_1'].notna() & df_raw['address_1'].notna()]

# Set 'city' column with correct dtype
df['city'] = pd.Series(dtype='object')

# Use vectorized operations instead of explicit loop
df['city'] = df['address_1'].str.split(',').str[-1].str.lstrip()

# Select columns
selected_columns = ['address_1', 'city', 'pricing_1', 'title', 'url']

# Filter rows based on 'pricing_1' condition
result = df[~df['pricing_1'].str.contains('Brak informacji o usługach i cenach', case=False, na=False)].dropna(subset=['pricing_1'])

# Display the result
result[selected_columns]
result.to_csv('output.csv', delimiter=';')
```
